```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/381a9472-ce83-407d-9a64-1b8c23af83df")
```

```python
df1
```

```python
date=df1["extract_date"]
R=df1["R"]
taux_pos=df1["tx_pos"]
taux_inc=df1["tx_incid"]
taux_occupation=df1["taux_occupation_sae"]
```

```python
param=[R, taux_pos, taux_inc, taux_occupation]
```

```python
max(taux_occupation)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
title=["Taux de reproduction", "Taux_positivité", "Taux_incidence", "Taux_occupation"]
for i in range(len(param)):   
    fig = px.line(x=date, y=param[i], title=title[i]+" en fonction du temps")
    # print(fig)
    fig.show()
```

***Etude du 2eme confinement*** 

30 Octobre 2020 -> 15 Décembre 2020

```python
confinement2=df1[df1["extract_date"].between("2020-10-20","2020-12-20")]
```

```python
confinement2
```

```python
date_conf2=confinement2["extract_date"]
R_conf2=confinement2["R"]
taux_pos_conf2=confinement2["tx_pos"]
taux_inc_conf2=confinement2["tx_incid"]
taux_occupation_conf2=confinement2["taux_occupation_sae"]
param_conf2=[R_conf2,taux_pos_conf2,taux_inc_conf2,taux_occupation_conf2]
```

```python
for i in range(len(param)):   
    fig = px.line(x=date_conf2, y=param_conf2[i], title=title[i]+" en fonction du temps")
    # print(fig)
    fig.show()
```

Vu l'augmentation du taux de reproduction (ce qui est équivaut au fait que l'épidémie se propage plus vite), du taux de positivité et du taux d'incidence,le gouvernement annonce un 2ème confinement à partir du 30/10/2020. le pic de ces variables a été atteint vers le 28 octobre.

***Remarque ***

* Le pic du taux d'occupation quant à lui est atteint vers le 15 novembre puisqu'il faut tenir compte du fait que le délai entre une mesure de contrôle (confinement, couvre feu...) et sa manifestation au niveau des nouvelles hospitalisations est d’environ 2 semaines (on estime à 5 jours le temps d’apparition des symptômes et à 8 jours le temps entre l’apparition des symptômes et l’hospitalisation pour les cas sévères).

* Le taux de positivité dépend du nombre de tests effectués.

-->La variable qui décrit le mieux l'évolution de l'émpidémie est le taux de reproduction


Dans les jours qui suivent l'annonce du confinement, on constate une diminution des taux de reproduction, de positivité et d'incidence. Le taux de reproduction est désormais inférieur à 1 (7 novembre 2020) --> l'épidémie régresse

Avec l'annonce de la première phase de déconfinement (28 novembre 2020), le taux de reproduction augmente et l'épidémie se développe à nouveau.

```python
for i in range(len(param_conf2)):   
    plt.figure()
    plt.boxplot(param_conf2[i])
    plt.title(title[i]+"_2ème_confinement")
    plt.show()
```

```python
df3=df2.copy()
df3=df3.groupby(["libelle_reg","extract_date"])["taux_occupation_sae"].mean()
```

```python
df3=df3.reset_index()
```

```python
df3=df3.pivot(index="extract_date", columns="libelle_reg", values="taux_occupation_sae")
```

```python
df3
```

```python
df3=df3.reset_index()
date=df3["extract_date"]
columns=df3.columns[1:]
```

```python
df3
```

```python
plt.figure(figsize=(15,7))
date=df3["extract_date"]
date=pd.to_datetime(date)
for i in range(len(columns[1:])):
  plt.plot(date,df3[columns[i]], label=columns[i])
plt.legend()
plt.xlabel("date")
plt.ylabel("taux_d'occupation")
plt.show()
```

En traçant l'evolution, pour les différentes régions, du taux d'incidence par exemple, on peut voir la différence qui existent entre ces région en terme de propagation de l'épidémie. Ceci permettrait de prendre des mesures régionales comme le fait d'annoncer un confinement régional.

***Vaccination***

Dans cette partie, on ne cosidère que les dates à partir du 27-12-2020 (date de début de la campagne de vaccination)

```python
vac=pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/b273cf3b-e9de-437c-af55-eda5979e92fc", sep=";")
```

```python
vac
```

```python
vac0=vac[vac['vaccin']==0]
```

```python
dose1=vac0["n_cum_dose1"]
dose2=vac0["n_cum_dose2"]
dose=[dose1,dose2]
```

```python
plt.figure(figsize=(15,6))
for i in range (2):
  plt.plot(dose[i], label="dose "+ str(i+1))
plt.legend()
plt.xlabel("date")
plt.ylabel("nombre de doses")
plt.title("Evolution de la compagne de vaccination")
plt.show()
```

```python
df1_effet_vac=df1[df1["extract_date"].between("2020-12-27", "2021-10-21	")]
df1_effet_vac["extract_date"]=pd.to_datetime(df1_effet_vac["extract_date"])
df1_effet_vac.set_index(keys="extract_date",drop=True)
vac0["jour"]=pd.to_datetime(vac0["jour"])
vac0.set_index(keys="jour",drop=True)
```

```python
fig=plt.figure(figsize=(15,8))
ax=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="2", frame_on=False)

date=df1_effet_vac["extract_date"]
ax.plot(list(vac0["jour"]),vac0["n_cum_dose1"], color="C0", label="dose1")
ax.plot(list(vac0["jour"]),vac0["n_cum_dose2"], color="C3", label="dose2")
ax.set_xlabel("jours", color="C0")
ax.set_ylabel("nombre de vaccin", color="C0")
# ax.tick_params(axis='x', colors="C0")
# ax.tick_params(axis='y', colors="C0")

ax2.plot(list(df1_effet_vac["extract_date"]),df1_effet_vac["taux_occupation_sae"], color="C1", label="taux d'occupation")
ax2.yaxis.tick_right() 
ax2.set_ylabel("taux d'occupation", color="C1")       
ax2.yaxis.set_label_position('right') 
# ax2.tick_params(axis='x', colors="C1")
# ax2.tick_params(axis='y', colors="C1")
ax.legend()
ax2.legend()
```

--> On voit bien l'effet de la vaccination sur l'évolution du taux d'occupation des hopitaux. La nouvelle vague de l'épidémie (à partir du fin du mois d'août) reste sans conséquences graves puisque le pic du taux d'occupation est de l'ordre de 40% comparé au 120% du 1er pic.
